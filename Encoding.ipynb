### Most Frequent Encoding

```python
import pandas as pd
```

```python
df=pd.DataFrame({"city":["Delhi","Mumbai","Delhi",None,"Mumbai","Delhi"]})
```

```python
#most frequently values nikalna
most_frequent=df["city"].mode()[0]
most_frequent
```

```python
#fill krna
df.fillna(most_frequent,inplace=True)
```

### Target Encoding

```python
import pandas as pd
```

```python
df=pd.DataFrame({"city":["Delhi","Mumbai","Delhi","chennai","Mumbai","Delhi"],
                "sale":[100,200,150,120,250,200]})
```

```python
#Target Encoding
mean_encoded=df.groupby("city")["sale"].mean()
mean_encoded
```

```python
df=df["city"].map(mean_encoded)
```

```python
df
```

### ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
import numpy as np

# Sample data
sizes = [['Small'], ['Medium'], ['Large'], ['Small'], ['Large']]

# Define the order
size_order = [['Small', 'Medium', 'Large']]

# Create ordinal encoder with category order
oe = OrdinalEncoder(categories=size_order)

# Fit and transform
encoded_sizes = oe.fit_transform(sizes)

print("Original:", sizes)
print("Encoded :", encoded_sizes)
```

### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

# Sample data
colors = ['Red', 'Blue', 'Green', 'Blue', 'Red']

# Create label encoder
le = LabelEncoder()

# Fit and transform
encoded_colors = le.fit_transform(colors)

print("Original:", colors)
print("Encoded :", encoded_colors)
```
