##### Types of Binning (Strategy)

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import KBinsDiscretizer

# Sample data
X = np.array([[1], [2], [3], [5], [8], [10], [13], [18], [21], [25], [30]])

# Binning strategies
strategies = ['uniform', 'quantile', 'kmeans']

# Plotting
plt.figure(figsize=(12, 4))

for i, strategy in enumerate(strategies, 1):
    kbd = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy=strategy)
    X_binned = kbd.fit_transform(X)
    
    plt.subplot(1, 3, i)
    plt.scatter(X.ravel(), [1]*len(X), c=X_binned.ravel(), cmap='viridis', s=100)
    plt.title(f'{strategy.capitalize()} Binning')
    plt.yticks([])
    plt.xlabel('Value')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

Output Interpretation:
##### Uniform Binning:

Sab bins ka width barabar hai, lekin har bin mein samples uneven hain.

##### Quantile Binning:

Har bin mein approx. same number of samples hain, lekin width unequal hai.

##### KMeans Binning:

Bins clusters ke basis pe banaye gaye hain, toh naturally grouped values ek bin mein aa jaate hain.
